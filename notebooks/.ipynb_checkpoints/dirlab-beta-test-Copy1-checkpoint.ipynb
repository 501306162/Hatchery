```julia
moving = readDIRLab("C:\\Users\\qyp\\Dropbox\\CTdata\\case5_T50-ssm.img", NRIRHOPM.DIRLab4DCT5);
```

```julia
originΔ = vecnorm(moving-fixed)
```

```julia
displacements32 = [SVector(i,j,k) for i in -6:3:6, j in -6:3:6, k in -6:3:6]
displacements42 = [SVector(i,j,k) for i in -4:2:4, j in -4:2:4, k in -4:2:4]
displacements64 = [SVector(i,j,k) for i in -2:2:2, j in -2:2:2, k in -2:2:2]
displacements85 = [SVector(i,j,k) for i in -2:1:2, j in -2:1:2, k in -2:1:2]
displacements128 = [SVector(i,j,k) for i in -1:1, j in -1:1, k in -1:1]
displacements256 = [SVector(i,j,k) for i in -1:1, j in -1:1, k in -1:1]

displacementSet = [displacements32, displacements42, displacements64, displacements85, displacements128, displacements256]
gridSet = [(32,32,13), (42,42,17), (64,64,26), (85,85,35), (128,128,53), (256,256,106)]

warpped, d, spec, energy = multilevel(fixed, moving, displacementSet, gridSet, MixHOPM(), SAD(), ones(6), 
    TAD(), 10*[1/3,1/2,1/2,1,1,1])
topologyΔ = vecnorm(warpped[end]-fixed)
```

```julia
displace = [SVector(i,j,k) for i in -1:1, j in -1:1, k in -1:1]
displacements32 = [SVector(i,j,k) for i in -6:6, j in -6:6, k in -6:6]
displacements42 = [SVector(i,j,k) for i in -5:5, j in -5:5, k in -5:5]
displacements64 = [SVector(i,j,k) for i in -4:4, j in -4:4, k in -4:4]
displacements85 = [SVector(i,j,k) for i in -3:3, j in -3:3, k in -3:3]
displacements128 = [SVector(i,j,k) for i in -2:2, j in -2:2, k in -2:2]
# displacementSet = [displacetopo, displacements32, displacements64, displacements128]
displacementSet = [displace]
gridSet = [(256,256,106)]
warpped, d, spec, energy = multilevel(fixed, moving, displacementSet, gridSet, MixHOPM(), SAD(), ones(1), TAD(), 0.5*[1])
topologyΔ = vecnorm(warpped[end]-fixed)
```
