```julia
using NRIRHOPM, FileIO, Images, Plots, Interpolations; gr()
```

```julia
include("draw2.jl")
```

```julia
circle = Gray.(load("./ring.png"))
```

```julia
square = Gray.(load("./square.png"))
```

```julia
size(circle), size(square), typeof(circle)
```

```julia
square, circle = restrict(square), restrict(circle); 
```

```julia
fixed, moving = square[1:64,1:64],  circle[1:64,1:64];
```

```julia
displacements0 = [SVector(i,j) for i in -4:4, j in -4:4]
displacements1 = [SVector(i,j) for i in -2:2, j in -2:2]
displacements2 = [SVector(i,j) for i in -1:1, j in -1:1]
displacementSet = [displacements0, displacements1, displacements2]
gridSet = [(30,30), (80,80), (160,160)]
originΔ = vecnorm(moving-fixed)
```

```julia
displacetopo = [SVector(i,j) for i in -6:6, j in -6:6]
# displacements = [SVector(i,j) for i in -3:3, j in -3:3]
# displacements2 = [SVector(i,j) for i in -1:1, j in -1:1]
displacementSet = [displacetopo]#, displacements, displacements2]
gridSet = [(64,64)] #,(64,64),(128,128)]
originΔ = vecnorm(moving-fixed)
```

```julia
# with topology preservation
warpped, d, spec, energy = multilevel(fixed, moving, displacementSet, gridSet, topology=NRIRHOPM.TP2D(), 
α=1, β=0.1, χ=10, method=MixHOPM(constraint=:vecnorm, maxIteration=20))
topologyΔ = vecnorm(warpped[end]-fixed)
```

```julia
colorview(Gray, warpped[1])
```

```julia
colorview(Gray, warpped[end])
```
